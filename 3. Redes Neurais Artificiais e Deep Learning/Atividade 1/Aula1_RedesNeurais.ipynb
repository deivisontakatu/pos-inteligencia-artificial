## 1. Contextualização do Problema
### Área ou setor analisado
O setor analisado é o **educacional**, com foco no **ensino online** em instituições de ensino.

### Justificativa da escolha da base de dados
A base foi escolhida por permitir a análise de fatores tecnológicos, socioeconômicos e educacionais que influenciam a adaptação dos estudantes ao ensino online.

### Problema, desafio ou oportunidade organizacional
Apesar da ampla adoção do ensino online, muitos estudantes apresentam apenas adaptação moderada, o que pode comprometer engajamento, satisfação e permanência.

## 2. Descrição da Base de Dados
- **Origem:** Base pública (Kaggle)
- **Tipo de dados:** Categóricos e qualitativos
- **Volume aproximado:** 1.205 registros e 14 variáveis

### Principais variáveis
- Gender
- Age
- Education Level
- Internet Type
- Device
- Financial Condition
- Adaptivity Level (variável-alvo)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['figure.figsize'] = (8,5)

df = pd.read_csv('students_adaptability_level_online_education.csv')
df.head()
```

## 3. Análise Exploratória dos Dados
A seguir são apresentadas estatísticas descritivas e visualizações gráficas para identificação de padrões.

```python
df.describe(include='all')
```

```python
df['Adaptivity Level'].value_counts().plot(kind='bar')
plt.title('Distribuição do Nível de Adaptabilidade')
plt.xlabel('Nível de Adaptabilidade')
plt.ylabel('Quantidade de Estudantes')
plt.show()
```

### Interpretação
Estudantes que utilizam predominantemente dispositivos móveis concentram-se nos níveis Low e Moderate de adaptabilidade, sugerindo limitações de usabilidade e experiência quando plataformas não são otimizadas para esse tipo de acesso.

## 4. Geração de Insights
- A maior parte dos estudantes apresenta nível de adaptabilidade moderado, indicando que conseguem acompanhar o ensino online, porém ainda enfrentam dificuldades que limitam uma adaptação plena. Esse padrão evidencia a necessidade de ações institucionais estruturadas para elevar a experiência educacional digital.
- Observa-se predominância do uso de dispositivos móveis, indicando que grande parte dos estudantes acessa o ensino online por meio de smartphones. Esse cenário reforça a necessidade de plataformas educacionais mobile-first, com interfaces responsivas e baixo consumo de dados.
- Estudantes que utilizam predominantemente dispositivos móveis concentram-se nos níveis Low e Moderate de adaptabilidade, sugerindo limitações de usabilidade e experiência quando plataformas não são otimizadas para esse tipo de acesso.
- Estudantes com condição financeira média ou baixa apresentam maior concentração em níveis reduzidos de adaptabilidade, indicando que fatores socioeconômicos influenciam diretamente a experiência no ensino online.

## 5. Propostas de Melhoria Organizacional
Com base nos padrões identificados na análise exploratória dos dados, são propostas as seguintes ações estratégicas para aprimorar a experiência do ensino online:

**1. Otimização de plataformas educacionais para dispositivos móveis**  
A predominância do uso de smartphones como principal meio de acesso ao ensino online evidencia a necessidade de adoção de uma abordagem *mobile-first*. Plataformas educacionais devem priorizar interfaces responsivas, navegação simplificada e baixo consumo de dados, visando melhorar a usabilidade e, consequentemente, a adaptação dos estudantes.

**2. Capacitação docente para o ensino digital**  
A concentração dos estudantes nos níveis de adaptabilidade *Low* e *Moderate* indica que, embora o ensino online seja funcional, há limitações pedagógicas que impactam a experiência de aprendizagem. A capacitação contínua dos docentes em metodologias digitais, design instrucional e uso de recursos interativos pode aumentar a percepção de utilidade do ensino online e elevar os níveis de adaptação discente.

**3. Ampliação do suporte técnico e pedagógico institucional**  
A baixa utilização de plataformas institucionais próprias e a diversidade de condições socioeconômicas dos estudantes reforçam a necessidade de fortalecimento do suporte técnico e pedagógico. Investimentos em atendimento tecnológico, orientação acadêmica e suporte contínuo aos estudantes tendem a reduzir barreiras de acesso e promover maior equidade no ensino online.







```python
# Percentual de estudantes que utilizam dispositivos móveis
mobile_usage = df['Device'].value_counts(normalize=True) * 100
mobile_usage
# Gráfico de apoio à decisão mobile-first
mobile_usage.plot(kind='bar')
plt.title('Uso de Dispositivos no Ensino Online')
plt.xlabel('Tipo de Dispositivo')
plt.ylabel('Percentual (%)')
plt.show()
```

```python
# Percentual de estudantes com adaptação não plena
non_high_adaptivity = df['Adaptivity Level'].value_counts(normalize=True) * 100
non_high_adaptivity
non_high_adaptivity.plot(kind='bar')
plt.title('Distribuição do Nível de Adaptabilidade dos Estudantes')
plt.xlabel('Nível de Adaptabilidade')
plt.ylabel('Percentual (%)')
plt.show()
```

```python
# Uso de LMS institucional
lms_usage = df['Self Lms'].value_counts(normalize=True) * 100
lms_usage
lms_usage.plot(kind='bar')
plt.title('Uso de LMS Institucional')
plt.xlabel('Utilização de LMS')
plt.ylabel('Percentual (%)')
plt.show()
# Tipo de internet utilizada
internet_usage = df['Internet Type'].value_counts(normalize=True) * 100
internet_usage
internet_usage.plot(kind='bar')
plt.title('Tipo de Acesso à Internet')
plt.xlabel('Tipo de Internet')
plt.ylabel('Percentual (%)')
plt.show()
```

## 6. Fundamentação Científica das Propostas
As propostas de melhoria organizacional apresentadas neste trabalho estão fundamentadas em estudos científicos consolidados que investigam a adaptação ao ensino online, a aceitação tecnológica e a permanência discente em ambientes virtuais de aprendizagem.

Estudos clássicos e contemporâneos apontam que a satisfação dos estudantes em cursos online é fortemente influenciada pela facilidade de uso e pela utilidade percebida das plataformas educacionais. Joo, Lim e Kim (2011) demonstram que esses fatores exercem impacto direto na satisfação e na persistência dos estudantes em ambientes virtuais de aprendizagem, indicando que plataformas intuitivas e bem estruturadas favorecem o engajamento e a continuidade no uso do ensino online.

Sob a perspectiva da aceitação tecnológica, Alharbi e Drew (2014) aplicam o Technology Acceptance Model (TAM) ao contexto educacional e evidenciam que a aceitação de sistemas de gestão da aprendizagem depende não apenas de aspectos técnicos, como usabilidade e desempenho, mas também do preparo pedagógico dos docentes. Os autores destacam que a capacitação docente é um elemento central para ampliar o uso efetivo das tecnologias educacionais, reforçando a importância de investimentos contínuos em formação para o ensino digital.

Complementarmente, Li et al. (2023) propõem o Adaptive Structural Learning and Technology Acceptance Model (ASL-TAM), demonstrando empiricamente que a adaptabilidade dos estudantes ao ensino online influencia positivamente a intenção de uso contínuo das plataformas educacionais. O estudo evidencia ainda que a satisfação com o ensino online atua como variável mediadora entre adaptabilidade e permanência, ressaltando a necessidade de ambientes educacionais adaptativos, aliados a suporte técnico e pedagógico institucional adequado.

Dessa forma, as propostas de otimização das plataformas educacionais, capacitação docente e ampliação do suporte institucional encontram respaldo teórico e empírico na literatura científica, evidenciando sua relevância e viabilidade para a melhoria da experiência educacional digital.
